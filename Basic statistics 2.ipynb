# Calculating pearson r

Pearson r is the coefficient of correlation between 2 variables.

If pearson r is over 0, then there is positive correlation between variables.

```python
from pydoc import help
from scipy.stats.stats import pearsonr
help(pearsonr)
```

```python
x = [2, 4, 1.5, 2, 3]
y = [7, 3, 8, 8, 6]

pearsonr(x,y)
```

# Graphing the relationship

As you see on the graph below there is positive correlation between the abscissa and the ordinate

```python
def f(x):
    return 0.66*x + 1.99

import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook


q = np.linspace(0,100,100)

plt.plot(q, map(f,q))
plt.show()
```

```python
pearsonr(q,map(f,q))
```
